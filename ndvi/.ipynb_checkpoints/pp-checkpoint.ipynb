```python
from scipy.ndimage.measurements import variance
import itertools

def lee_filter_scalar(img, size):
    img = img.astype(np.float32)
    img_mean = cv2.bilateralFilter(img, 10, size, size)
    img_sqr_mean = cv2.bilateralFilter(img**2, 10, size, size)
    img_variance = img_sqr_mean - img_mean**2
    
    overall_variance = variance(img)

    img_weights = img_variance / (img_variance + overall_variance)
    img_output = img_mean + img_weights * (img - img_mean)
    return img_output

cube['VH_filt'] = xr.DataArray([lee_filter_scalar(ds.values, 3) for t, ds in cube['VH'].groupby('time')], dims=['time', 'lat', 'lon'])
```

```python
for i in range(1):
    org = cube['VH'].isel(time=i).values
    plt.imshow(np.log10(org), aspect='auto', cmap='gray')
    plt.title('Without Filtering')
    plt.show()

    arr = np.log10(np.abs(lee_filter_scalar(org, 3)))
    plt.imshow(arr, aspect='auto', cmap='gray')

    plt.title('With Lee Filtering')
    plt.show()
```

```python
def threshold(da):
    vals = da.values.flatten()
    vals = vals[np.isfinite(vals)]
    try:
        res = skimage.filters.threshold_minimum(vals)
    except RuntimeError:
        print('RuntimeError')
        res = 0
    return res

v = 'VH_filt'
vlog10 = v + '_log10'
cube[vlog10] = np.log10(cube[v])

da = cube[vlog10].isel(time=3)
print(threshold(da))
vals = da.values.flatten()
vals = vals[np.isfinite(vals)]

plt.hist(vals, bins=60)
plt.scatter(threshold(da), 50, color='r', s=100, zorder=3)
plt.grid()
plt.show()

cube['thresholds'] = xr.DataArray([threshold(ds) for t, ds in cube['VH_filt_log10'].groupby('time')], dims=['time'])
cube['water'] = cube[vlog10] < cube['thresholds']

da_water = cube['water'].where(cube['thresholds'] < -2, drop=True)

np.log10(da_water.mean(dim='time')).plot(robust=True)
plt.show()
```
