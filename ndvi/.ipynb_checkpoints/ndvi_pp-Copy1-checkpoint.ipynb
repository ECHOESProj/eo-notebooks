#### Importing dependancies
The user needs to first import the dependancies.

```python
# xcube_sh imports
from xcube_sh.cube import open_cube
from xcube_sh.config import CubeConfig
from xcube_sh.observers import Observers
from xcube_sh.viewer import ViewerServer

# xcube imports
from xcube.core.maskset import MaskSet
from xcube.core.geom import mask_dataset_by_geometry
from xcube.core.geom import clip_dataset_by_geometry

# Various utilities
import json
from datetime import date, timedelta, datetime
import xarray as xr
import shapely.geometry
import IPython.display
import zarr
import numpy as np
import matplotlib.pyplot as plt
```

```python
bbox = (-3.93, 52.56, -3.9, 52.58)
IPython.display.GeoJSON(shapely.geometry.box(*bbox).__geo_interface__)
```

 #### Setting the spatial resolution of the cube
 The next step is to specify the spatial resolution (pixel size) on which the raster data (in this case Sentinel-2 imageries) should be resampled and stored in.
 
In this case, the desired resolution is roughly 10 meters per pixel:


```python
time_range = ['2017-01-01', '2020-12-31']
spatial_res = 0.00018   # = 10.038 meters in degree>
cube_config = CubeConfig(dataset_name='S2L2A',
                         band_names=['B04', 'B08', 'CLM'],
                         tile_size=[512, 512],
                         bbox=bbox,
                         spatial_res=spatial_res,
                         time_range=time_range,
                         time_period='1D')

request_collector = Observers.request_collector()
cube = open_cube(cube_config, observer=request_collector)
```

```python
ndvi = ((cube['B08'] - cube['B04']) / (cube['B08'] + cube['B04'])).where(np.logical_not(cube.CLM))
def cube_by_year_month():
    for year, ds_year in ndvi.groupby('time.year'):
        for month, ds_month in ds_year.groupby('time.month'):
            ds = ds_month.mean('time')
            ds.expand_dims(dim='month')
            ds['month'] = datetime(year, month, 15)
            yield ds
```

```python
ds = xr.concat([ds for ds in cube_by_year_month()], dim='month')
```

```python
ds.isel(month=0).plot.imshow(cmap='YlGn')
plt.show()
```

```python
ndvi_season = ndvi.groupby('time.season').mean(dim='time').reindex(season=["DJF", "MAM", "JJA", "SON"])
for ds in ndvi_season:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    ds.plot(cmap='YlGn', vmin=0.2, vmax=0.9, ax=ax1)
    ds.plot.hist(ax=ax2, bins=60)
    plt.show()
```

```python
ndvi_year = ndvi.groupby('time.year').mean(dim='time')
ds = ndvi_year.mean(dim='lat').mean(dim='lon')
ds.plot()
plt.grid()
plt.show()
```

```python
for y, ds in ndvi.groupby('time.year'):
    ds.mean(dim='time').plot.imshow(cmap='YlGn', vmin=0.2, vmax=0.9)
    plt.title(y)
    plt.show()
```

```python
a, b = 2020, 2019
ndvi_a = ndvi_year.sel(year=a)
ndvi_a -= np.median(ndvi_a)
ndvi_b = ndvi_year.sel(year=b)
ndvi_b -= np.median(ndvi_b)
(ndvi_a - ndvi_b).plot(robust=False)
plt.show()
```

```python
ds = ndvi.groupby('time.season').mean(dim='time').mean(dim='lat').mean(dim='lon').reindex(season=["DJF", "MAM", "JJA", "SON"])
plt.plot(ds.values)
plt.xticks(range(4), ds.season.values)
plt.grid()
plt.show()
```
