# NDVI

The NDVI is a dimensionless index that describes the difference between visible and near-infrared reflectance of vegetation cover and can be used to estimate the density of green on an area of land (Weier and Herring, 2000)

```python
# xcube_sh imports
from xcube_sh.cube import open_cube
from xcube_sh.config import CubeConfig
from xcube_sh.observers import Observers
from xcube_sh.viewer import ViewerServer

# xcube imports
from xcube.core.maskset import MaskSet
# from xcube.core.geom import mask_dataset_by_geometry
# from xcube.core.geom import clip_dataset_by_geometry

# Various utilities
import json
from datetime import date, timedelta, datetime
import xarray as xr
import shapely.geometry
import IPython.display
import zarr
import numpy as np
import matplotlib.pyplot as plt
```

```python
bbox = (-3.93, 52.56, -3.9, 52.58)
IPython.display.GeoJSON(shapely.geometry.box(*bbox).__geo_interface__)
```

```python
time_range = ['2017-01-01', '2022-12-31']
spatial_res = 0.00018   # = 10.038 meters in degree>
cube_config = CubeConfig(dataset_name='S2L2A',
                         band_names=['B04', 'B08', 'CLM'],
                         tile_size=[512, 512],
                         bbox=bbox,
                         spatial_res=spatial_res,
                         time_range=time_range,
                         time_period='1D')

request_collector = Observers.request_collector()
try:
    cube = xr.open_zarr('data/cube_pp.zarr')
except zarr.errors.GroupNotFoundError:
    cube = open_cube(cube_config, observer=request_collector)
    cube.to_zarr('data/cube_pp.zarr')
```

#### Compute the NDVI

```python
ndvi = ((cube['B08'] - cube['B04']) / (cube['B08'] + cube['B04'])).where(np.logical_not(cube.CLM))
```

#### Plot by Season

```python
ndvi_season = ndvi.groupby('time.season').mean(dim='time').reindex(season=["DJF", "MAM", "JJA", "SON"])
for ds in ndvi_season:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    ds.plot(cmap='YlGn', vmin=0.2, vmax=0.9, ax=ax1)
    ds.plot.hist(ax=ax2, bins=60)
    plt.show()
```

```python
ds = ndvi.groupby('time.season').mean(dim='time').mean(dim='lat').mean(dim='lon').reindex(season=["DJF", "MAM", "JJA", "SON"])
plt.plot(ds.values)
plt.xticks(range(4), ds.season.values)
plt.grid()
plt.show()
```

As expected, the NDVI is greatest in the summer and at a minimium in the winter

#### Plot by Year

There is not much difference year-to-year in the mean NDVI

```python
ndvi_year = ndvi.groupby('time.year').mean(dim='time')
ds = ndvi_year.mean(dim='lat').mean(dim='lon')
ds.plot()
plt.grid()
plt.show()
```

```python
for y, ds in ndvi.sel(time=slice('2017-01-01', '2020-12-31')).groupby('time.year'):
    ds.mean(dim='time').plot.imshow(cmap='YlGn', vmin=0.2, vmax=0.9)
    plt.title(y)
    plt.show()
```

#### Compare Years

```python
a, b = 2020, 2019
ndvi_a = ndvi_year.sel(year=a)
ndvi_b = ndvi_year.sel(year=b)
(ndvi_a - ndvi_b).plot(robust=False)

#p_lon, p_lat = -3.925, 52.573189 
p_lon, p_lat = -3.908, 52.573 # deforested
plt.scatter([p_lon], [p_lat], s=10, c='y')

plt.show()

ndvi.sel(lat=p_lat, lon=p_lon, method='nearest').resample(time='1D').mean().dropna(dim='time').to_series().plot()
plt.grid()
plt.show()
```

```python
years = []
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))
for y, dsyear in ndvi.sel(time=slice('2017-01-01', '2022-12-31')).groupby('time.year'):
    years.append(y)
    ds = dsyear.mean(dim='lat').mean(dim='lon').dropna(dim='time')
    dsr = ds.rolling(time=10).mean()
    
    ds_count = dsyear.count(dim='lat').mean(dim='lon').dropna(dim='time')
    dsr_count = ds_count.rolling(time=10).mean()
        
    ax1.plot(dsr_count.time.dt.dayofyear, dsr_count)
    ax2.plot(dsr.time.dt.dayofyear, dsr)
plt.xlabel('Day of Year')
plt.ylabel('NDVI')
plt.grid()
plt.legend(years)
plt.show()
```
